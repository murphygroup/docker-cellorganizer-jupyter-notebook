```python
import sys
import os
import ipywidgets as ui
from IPython.display import display, clear_output
from cellorganizer.tools import imshow

files = []

class PathSelector():

    def __init__(self, start_dir, select_file=True):
        self.file        = [] 
        self.select_file = select_file
        self.start_dir = start_dir
        self.cwd         = start_dir
        self.select      = ui.SelectMultiple(options=['init'],value=[],rows=10,description='', disabled=False) 
        self.accord      = ui.Accordion(children=[self.select]) 
        self.confirm     = ui.Button(description='Done', disabled=False, 
                                     button_style='', # 'success', 'info', 'warning', 'danger' or ''
                                     tooltip='Click me',
                                     icon='check')

        self.accord.selected_index = 0 # Start closed (showing path only)
        self.refresh(self.cwd)
        
        self.select.observe(self.on_update,'value')
        self.confirm.on_click(self.confirmed)
        
    
    

    def on_update(self,change):
        if len(change['new']) > 0:
            self.file = []
            for i in range(1, len(change['new'])):
                self.file.append(os.path.abspath(os.path.join(self.cwd, change['new'][i])))
            self.refresh(change['new'][0])      
        print(self.file)
        
    def confirmed(self, b):
        global files
        files = self.file
        self.__init__(self.start_dir, True)
        clear_output()
        display(self.accord)
        display(self.confirm) 

    def refresh(self, item):
        path = os.path.abspath(os.path.join(self.cwd,item))
        
        if os.path.isfile(path):
            if self.select_file:
                self.file.append(path)
                self.accord.selected_index = 0
            else:
                self.select.value = []

        else: # os.path.isdir(path)
            self.file = [] 
            self.cwd  = path

            # Build list of files and dirs
            keys = ['[..]']; 
            for item in os.listdir(path):
                if item[0] == '.':
                    continue
                elif os.path.isdir(os.path.join(path,item)):
                    keys.append('['+item+']'); 
                else:
                    keys.append(item); 

            # Sort and create list of output values
            keys.sort(key=str.lower)
            vals = []
            for k in keys:
                if k[0] == '[':
                    vals.append(k[1:-1]) # strip off brackets
                else:
                    vals.append(k)

            # Update widget
            self.select.options = list(zip(keys,vals)) 
            with self.select.hold_trait_notifications():
                self.select.value = []

f = PathSelector('./')
display(f.accord)
display(f.confirm)
```
